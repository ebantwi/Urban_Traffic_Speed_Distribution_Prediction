```python
import pandas as pd, numpy as np, warnings

warnings.filterwarnings('ignore')
pd.set_option('display.max_columns', None)
```

```python
#interchange segments
interchange_segs = pd.read_csv('freeway_interchange_segs.csv')
interlist = interchange_segs['OID_'].tolist()
interchange_segs.head()
```

```python
#all netscreen interchange segments
freeway_segs = pd.read_csv('freeway_with_ramps.csv')
freeway_segs.head()
```

```python
#horizontal curvature 
freeway_segs['DegHCurve_LWA'] = ((freeway_segs['CU_A_LENGTH']*1.7)+(freeway_segs['CU_B_LENGTH']*4.45)+
                          (freeway_segs['CU_C_LENGTH']*6.95)+(freeway_segs['CU_D_LENGTH']*11.2)+
                          (freeway_segs['CU_E_LENGTH']*20.95)+(freeway_segs['CU_F_LENGTH']*30))/(freeway_segs['SECTION_LENGTH'])
```

```python
#vertical grade
freeway_segs['DegGrade_LWA'] = ((freeway_segs['GR_A_LENGTH']*0.2)+(freeway_segs['GR_B_LENGTH']*1.45)+
                          (freeway_segs['GR_C_LENGTH']*3.45)+(freeway_segs['GR_D_LENGTH']*5.45)+
                          (freeway_segs['GR_E_LENGTH']*7.45)+(freeway_segs['GR_F_LENGTH']*9.45))/(freeway_segs['SECTION_LENGTH'])
```

```python
freeway_segs['TRUCK_PERC'] = freeway_segs['AVG_COMBINATION'] + freeway_segs['AVG_SINGLE_UNIT']
```

```python
#not in interchange segments
segs = freeway_segs.loc[~freeway_segs.OID_.isin(interlist)].reset_index(drop=True)
segs.head()
```

```python
ln = segs.sort_values(by=['ROUTE_ID', 'BEGIN_POINT', 'END_POINT'], ascending = True).reset_index(drop=True)
ln['LNID'] = ln.index
ln.head()
```

```python
print(len(ln))
ln.tail()
```

```python
#links aggregation
revid = 1
i = 0
while i < len(ln)-1:
    ln.loc[i, 'LNSegID'] = revid
    if ln.loc[i, 'ROUTE_ID'] == ln.loc[i+1, 'ROUTE_ID']: # new segment if either of these changes
        if (ln.loc[i, 'AADT'] != ln.loc[i+1, 'AADT']) |\
           (ln.loc[i, 'PEAK_CAPACITY_HCM6'] != ln.loc[i+1, 'PEAK_CAPACITY_HCM6']) |\
           (ln.loc[i, 'THROUGH_LANES'] != ln.loc[i+1, 'THROUGH_LANES']) |\
           (ln.loc[i, 'TYPE_TERRAIN'] != ln.loc[i+1, 'TYPE_TERRAIN']) |\
           (ln.loc[i, 'LANE_WIDTH'] != ln.loc[i+1, 'LANE_WIDTH']) |\
           (ln.loc[i, 'SHLD_WIDTH_L'] != ln.loc[i+1, 'SHLD_WIDTH_L']) |\
           (ln.loc[i, 'SHLD_WIDTH_R'] != ln.loc[i+1, 'SHLD_WIDTH_R']) |\
           (ln.loc[i, 'SPEED_LIMIT_LWA'] != ln.loc[i+1, 'SPEED_LIMIT_LWA']) |\
           (ln.loc[i, 'MEDIAN_TYPE'] != ln.loc[i+1, 'MEDIAN_TYPE']) |\
           (ln.loc[i, 'MEDIAN_WIDTH'] != ln.loc[i+1, 'MEDIAN_WIDTH']) |\
           (ln.loc[i, 'RURAL_URBAN'] != ln.loc[i+1, 'RURAL_URBAN']) |\
           (ln.loc[i, 'TRUCK_PERC'] != ln.loc[i+1, 'TRUCK_PERC']):
            revid += 1
    else:
        revid += 1
    i += 1
ln.loc[i, 'LNSegID'] = revid
```

```python
ln.tail()
```

```python
first = lambda x: x.iloc[0] #use for fields that are same, select one
lwa = lambda x: np.average(x, weights=ln.loc[x.index, "SECTION_LENGTH"])

ln_grouped = ln.groupby('LNSegID', as_index=False).agg({'ROUTE_ID':first,
                                          'BEGIN_POINT': 'min',
                                          'END_POINT': 'max',
                                          'FS_SYSTEM' : first,
                                          'RURAL_URBAN':first, 
                                          'AT_GRADE_OTHER':'sum',
                                          'AT_GRADE_SIGNALS':'sum',    
                                          'AT_GRADE_SIGNS':'sum',              
                                          'DegHCurve_LWA':lwa,
                                          'DegGrade_LWA':lwa,
                                          'AADT': first,
                                          'DirAADT': first,
                                          'PEAK_CAPACITY_HCM6':first,              
                                          'TRUCK_PERC': first,
                                          'TYPE_TERRAIN':first,
                                          'THROUGH_LANES': first,
                                          'LANE_WIDTH': first,
                                           'SHLD_WIDTH_L':first,
                                           'SHLD_WIDTH_R':first,
                                           'SPEED_LIMIT_LWA':first,
                                           'MEDIAN_TYPE':first,
                                           'MEDIAN_WIDTH':first,
                                          })
```

```python
ln_grouped["SECTION_LENGTH"] =  abs(ln_grouped['END_POINT'] - ln_grouped['BEGIN_POINT'])
ln_grouped["INTERCHANGE_SEG"] = 'No'
print(f"Total segments: {len(ln_grouped)}")
ln_grouped['LNSegID'] = ln_grouped['LNSegID'].astype(int)
ln_grouped.tail()
```

```python
ln_grouped.head()
```

```python
#interchange segments to append to grouped non-interchange segments
intsegs = freeway_segs.loc[freeway_segs.OID_.isin(interlist)].reset_index(drop=True)
intsegs["INTERCHANGE_SEG"] = 'Yes'
intsegs.head()
```

```python
start = ln_grouped['LNSegID'].max() + 1
end = (ln_grouped['LNSegID'].max() + 1) + (len(intsegs))
intsegs['LNSegID'] = list(range(int(start), int(end)))
intsegs.tail()
```

```python
intsegs_filtered = intsegs[ln_grouped.columns]
intsegs_filtered.head()
```

```python
## append intsegs to ln_grouped
final_segs = pd.concat([ln_grouped, intsegs_filtered], ignore_index=True)
final_segs.rename(columns = {'BEGIN_POINT':'BEGIN_MP', 'END_POINT':'END_MP'}, inplace=True)
final_segs.tail()
```

```python
final_segs.to_parquet('final_segs.parquet', index=False)
```

### Conflation Table

```python
cols = ['ROUTE_ID', 'BEGIN_POINT', 'END_POINT', 'AllRds_Dir', 'LinkDir']
ct = pd.read_csv('conf_table.csv', usecols=cols)
ct.head()
```

### Overlay

```python
def overlay(df_roads, df_conf):
    rds_overlay = pd.merge(df_roads,
                           df_conf, 
                           left_on = ['ROUTE_ID'],
                           right_on = ['ROUTE_ID'])
    
    
    rds_overlay = rds_overlay[(rds_overlay['BEGIN_MP']<rds_overlay['END_POINT']) & (rds_overlay['END_MP']>rds_overlay['BEGIN_POINT'])]
    rds_overlay['bmp'] = rds_overlay[['BEGIN_MP','BEGIN_POINT']].max(axis=1)
    rds_overlay['emp'] = rds_overlay[['END_MP','END_POINT']].min(axis=1)
    return rds_overlay
```

```python
#overlay both directions
ov = overlay(final_segs, ct)
ov.head()
```

### Read in 2018 Speed

```python
sp8 = pd.read_parquet('freeway_spd18.parquet', columns=['LINK-DIR','DATE-TIME', 'EPOCH-60MIN', 'LENGTH','MEAN'])
sp8.columns = ['LinkDir','DateTime', 'Hour','Here_Length', 'Mean']
sp8['Mean'] = round(sp8['Mean']/1.609, 2) #kph to mph
sp8.head()
```

```python
#takes longer to run. save intermediate results in next cell
m8 = pd.merge(ov, sp8, on='LinkDir', )
m8['SegLength'] = abs(m8['emp'] - m8['bmp'])
m8['MeanTT'] = m8['SegLength']/m8['Mean']
print(len(m8))
m8.head()
```

```python
m8agg = m8.groupby(['LNSegID','DateTime','Hour'], as_index=False).agg({'MeanTT':'sum', 'SegLength':'sum'})
m8agg['MeanSpeed'] = m8agg['SegLength']/m8agg['MeanTT']
m8agg.head()
```

```python
m8agg.to_parquet('aggregated_freeway_speeds_18.parquet', index=False)
```
